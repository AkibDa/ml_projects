```python
import numpy as np
import pandas as pd
```

```python
breast = pd.read_csv('data.csv')
breast.head()
```

```python
breast['diagnosis'].value_counts()
```

```python
breast.drop(['Unnamed: 32'], axis=1, inplace=True)
breast.head()
```

```python
# from sklearn.preprocessing import LabelEncoder
# le = LabelEncoder()
```

```python
breast['diagnosis'] = breast['diagnosis'].map({'M': 1, 'B': 0})
breast.head()
```

```python
x = breast.drop(['id', 'diagnosis'], axis=1)
y = breast['diagnosis']
x.head()
y.head()
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
x_train.shape, x_test.shape, y_train.shape, y_test.shape
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
x_train = sc.fit_transform(x_train)
```

```python
x_test = sc.transform(x_test)
x_train.shape, x_test.shape
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
```

```python
models = {
  "Logistic Regression": LogisticRegression(),
  "K-Nearest Neighbors": KNeighborsClassifier(),
  "Support Vector Machine": SVC(),
  "Decision Tree": DecisionTreeClassifier(),
  "Random Forest": RandomForestClassifier()
}
```

```python
from sklearn.metrics import accuracy_score

model_accuracy = {}

for name, model in models.items():
  model.fit(x_train, y_train)
  
  y_pred = model.predict(x_test)
  
  accuracy = accuracy_score(y_test, y_pred)
  
  model_accuracy[name] = accuracy
  print(f"{name} Accuracy: {accuracy:.4f}")
```

```python
best_model_name = max(model_accuracy, key=model_accuracy.get)
best_accuracy = model_accuracy[best_model_name]

print("\n------------------------------------")
print(f"The best performing model is: {best_model_name}")
print(f"It achieved an accuracy of: {best_accuracy:.4f}")
print("------------------------------------")
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score

results_list = []

for name, model in models.items():
  model.fit(x_train, y_train)
  
  y_pred = model.predict(x_test)
  
  accuracy = accuracy_score(y_test, y_pred)
  precision = precision_score(y_test, y_pred)
  recall = recall_score(y_test, y_pred)
  f1 = f1_score(y_test, y_pred)
  
  results_list.append({
    "Model": name,
    "Accuracy": accuracy,
    "Precision": precision,
    "Recall": recall,
    "F1-Score": f1
  })
```

```python
results_df = pd.DataFrame(results_list)

results_df.set_index('Model', inplace=True)

results_df = results_df.round(4)

print("Model Performance Comparison:")
print(results_df)
```
